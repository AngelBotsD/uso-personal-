```python
#@title Validar Colab
import requests
from IPython.display import display, Markdown

def obtener_bandera_emoji(codigo_pais):
    """Genera el emoji de la bandera basado en el cdigo ISO del pas"""
    if not codigo_pais:
        return ""
    return "".join([chr(ord(c) + 127397) for c in codigo_pais.upper()])

def validar_acceso_colab():
    # 1. Tu lista blanca
    paises_permitidos = [
        "Australia", "Austria", "Belgium", "Brazil", "China", "Finland",
        "France", "Germany", "Hong Kong", "Indonesia", "Israel", "Italy",
        "Japan", "Malaysia", "Netherlands", "Poland", "Portugal",
        "Saudi Arabia", "Singapore", "Spain", "Sweden", "Switzerland",
        "Taiwan", "Thailand", "Turkey", "United Arab Emirates",
        "United Kingdom", "United States", "Vietnam"
    ]

    try:
        # Peticin a la API
        response = requests.get('http://ip-api.com/json/')
        data = response.json()

        # Datos
        ip = data.get('query')
        pais = data.get('country')
        codigo = data.get('countryCode')
        isp = data.get('isp')
        bandera = obtener_bandera_emoji(codigo)

        # 2. Validacin y Visualizacin para Colab
        if pais in paises_permitidos:
            # Mensaje de XITO en formato Markdown (Verde/Negrita)
            mensaje = f"""
###  ACCESO AUTORIZADO
> **Pas detectado:** {pais} {bandera}
> **IP:** `{ip}`
> **Estado:** El proceso puede continuar.
            """
            display(Markdown(mensaje))
            return True
        else:
            # Mensaje de ERROR en formato Markdown
            mensaje = f"""
###  ACCESO DENEGADO
> **Pas detectado:** {pais} {bandera}
> **IP:** `{ip}`
> **Razn:** Tu ubicacin no est en la lista de pases permitidos.
            """
            display(Markdown(mensaje))
            return False

    except Exception as e:
        print(f"Error de conexin: {e}")
        return False

# Ejecutar en la celda
validar_acceso_colab()
```
